```python
init = [".#./..#/###"]
```

```python
def flip(u):
    return "/".join(map(lambda x : x[::-1], u.split("/")))
def rotate(u):
    t = u.split("/")
    tt = [[t[len(t[0])-1-i][len(t[0])-1-j] for i in range(len(t[0]))] for j in range(len(t[0]))]
    return "/".join(["".join(tt[i]) for i in range(len(t[0]))])
def flip_rotate(u, n):
    #print u,n
    if n%2 == 0:
        return flip(u)
    else:
        return rotate(u)
def find(u, u1,n=0):
    #print u, len(u1.keys()[0])
    if u in u1:
        return u1[u]
    return find(flip_rotate(u,n+1),u1,n+1)
def find_map(u):
    #print 'findmap', u
    if len(u) == 5:
        return find(u,u2)
    else:
        return find(u,u3)
def convert(u):
    r = []
    for i in u:
        r.append(find_map(i))
    return r
```

```python
import numpy as np
def reformat(u):
    # u is a list of such strings
    # these strings are restacked to form the larger square
    n = len(u) # must be a square
    n = int(np.sqrt(n))
    t = []
    k = len(u[0].split("/")[0])
    for i in range(n):
        for l in range(k):
            t.append("")
    for i in range(n):
        for j in range(n):
            f = u[j+i*n].split("/")
            for p in range(k):
                t[i*k+p] += (f[p])
    return t
def unstack(u):
    k = 2 if len(u[0])%2 == 0 else 3
    r = []
    n = len(u[0])/k
    for i in range(n):
        t = [u[i*k+v] for v in range(k)]
        for j in range(n):
            tt = "/".join([t[v][j*k:(j+1)*k] for v in range(k)])
            r.append(tt)
    return r
```

```python
unstack(reformat(convert(init)))
```

```python
def move(u):
    return unstack(reformat(convert(u)))
```
